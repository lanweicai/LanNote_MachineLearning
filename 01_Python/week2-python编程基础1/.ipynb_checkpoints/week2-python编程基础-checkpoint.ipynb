````python
class Hero(People):
    def __init__(self,name, hp, damage, country):
        People.__init__(self, name, hp, damage)
        self.country = country
        
    def get_inf(self):  #self
        """"""
        print("{}".format(self.name))
        print("{}".format(self.damage))
        print("{}".format(self.hp))
        print("{}".format(self.country))   
        
    def attack(self,enemy):
        """selfenemy
           enemy
           enemyenemy - self
           """
        print(self.name)
        print("{}".format(self.damage))
        print("{}{}".format(enemy.name, enemy.hp))
        print("{}{}".format(self.name,enemy.name))
        enemy.hp-=self.damage
        print("{}{}".format(enemy.name, enemy.hp))
        print("***********************************")
````

````python
#
liufeng = Hero('', 4, 2, '')
ganning = Hero('', 4, 2, '')
````

````python
#ganning
ganning.get_inf()
print('#################')
#liufengganning
liufeng.attack(ganning)
#ganning
ganning.get_inf()
````
