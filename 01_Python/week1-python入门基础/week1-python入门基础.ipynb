````python
def continue_mul(start,end):
    '''连续整数相乘
       start：最小值
       end：最大值
       返回值：start到end之间连续整数的乘积
       '''
    x = 1
    for i in range(start,end+1):
        x = x*i
    return x
````

if 足够支付，输出‘包治百病’，及余额

if 不够支付，输出‘你的想象力已被限制’及差值

````python
price = 6999
money = input('输入花呗余额：')
money = float(money)
if money > price:
    print('包治百病,余额为：{}'.format(money-price))
if money < price:
    print('你的想象力被限制在{}元'.format(price-money))
if money = price:
    print('轻松就是快乐！')
````
